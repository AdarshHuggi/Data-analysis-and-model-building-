```python
#importing the librairs

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing the data set

df_2015= pd.read_csv('2015.csv')
df_2016= pd.read_csv('2016.csv')
df_2017= pd.read_csv('2017.csv')
```

Observation:
    2017 dataset not have Region columan and have whisker.high , whisker.low

```python
#Adding the Region column to data set 2017

countries = df_2015.Country.unique() 
country_region = {}
for country in countries:
  country_region[country] = df_2015[df_2015.Country == country]["Region"].to_string()[3:].strip()

```

```python
country_reg_df = pd.DataFrame(list(country_region.items()), columns=["Country", "Region"])
country_reg_df.head()
```

```python
df_2017 = pd.merge(df_2017, country_reg_df)
```

```python
df_2017.head()
```

```python
#Droping the columns

df_2015 = df_2015.drop("Standard Error", axis=1)
df_2016 = df_2016.drop(["Lower Confidence Interval", "Upper Confidence Interval"], axis=1)
df_2017 = df_2017.drop(["Whisker.high", "Whisker.low"], axis= 1)
```

```python
df_2015.columns
```

```python
# reordering columns in 2017 dataset
cols_reorder= ['Country', 'Region','Happiness.Rank', 'Happiness.Score',
       'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.',
       'Freedom', 'Generosity', 'Trust..Government.Corruption.',
       'Dystopia.Residual']
df_2017 = df_2017[cols_reorder]


# changing column names 
new_col_names = ['country', 'region', 'happiness_rank', 'happiness_score',
       'economy', 'family', 'health',
       'freedom', 'trust', 'generosity',
       'dystopia_residual']


df_2015.columns= new_col_names
df_2016.columns= new_col_names
df_2017.columns= new_col_names
```

```python
# adding "Year" column 
df_2015["Year"] = ['2015'] * df_2015.shape[0]
df_2016["Year"] = ['2016'] * df_2016.shape[0]
df_2017["Year"] = ['2017'] * df_2017.shape[0]
```

```python
df_2016.head()
```

```python
#Merging all three year the data
all_dataset =pd.concat([df_2015,df_2016,df_2017], ignore_index=True)
```

# Task 4 : Countries that became unhappy between 2015 to 2017

```python
combined_df = wh_df.pivot_table(index = 'Year', values='happiness_score', aggfunc=np.mean)
combined_df.head()
```

```python
col_2015 = df_2015[['country', 'happiness_score']]
col_2017 = df_2017[['country', 'happiness_score']]
combined_2015_17 = pd.merge(left = col_2015, right=col_2017, how='left', on='country')
combined_2015_17.head()
```

```python
decreased_happiness = combined_2015_17[combined_2015_17['happiness_score_x'] > combined_2015_17['happiness_score_y']]
decreased_happiness.head()
```

# Task 5 : Find the country whose happiness decreased by the most amount

```python
dec_happiness = decreased_happiness.copy()
dec_happiness['happiness_decreased_amt'] = decreased_happiness.loc[:,'happiness_score_x'] - decreased_happiness.loc[:,'happiness_score_y']
dec_happiness.head() 

dec_happiness.sort_values(by='happiness_decreased_amt', ascending=False).head(1)
```

```python
df  = dec_happiness.sort_values(by='happiness_decreased_amt', ascending=False).head()[['country', 'happiness_decreased_amt']]
df.plot(kind='bar', x='country', y = 'happiness_decreased_amt', rot=30)
```

#### "Venezuela" is the country whose happiness decreased by the most amount. by 1.56
